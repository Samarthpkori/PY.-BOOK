```python
vec = np.array([[-3],[2]])
origin = np.zeros(vec.shape) # origin point

plt.figure(figsize=(6,6))
plt.quiver(*origin, *vec, color=['brown'], scale=1, units='xy')

plt.grid()
plt.xlim(-5,5)
plt.ylim(5,-5)
plt.gca().set_aspect('equal')
plt.show()
```
